#### Maxwellâ€“Boltzmann Velocity Distribution
In three dimensions, the LTE velocity distribution is  
$$
f_{\mathrm{MB}}(v;T)
=
4\pi
\left( 
  \frac{m}
       {2\pi k_{\mathrm B}T} \right)^{3/2}
v^2
\exp\!\left( -\frac{mv^2}{2k_{\mathrm B}T} \right),
$$
where 
- $m^*$ is the carrier effective mass, and 
- $T$ is the local temperature.
$\blacksquare$


```python
from solidstate import ConstantsSI
import numpy as np


def mb_velocity_distribution(
    m: float,
    T: float,
    v: float,
    k_B: float = ConstantsSI.k_B
):
  assert isinstance(v, float) or isinstance(v, np.ndarray)

  f_mb = 4 * np.pi
  f_mb *= (m/(2*np.pi*k_B))
```
