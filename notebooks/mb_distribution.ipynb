# Maxwell Boltzmann Distribution

**Definition (Maxwellâ€“Boltzmann distribution).**  
For a classical particle in thermal equilibrium at temperature $T$, the energy distribution is  
$$
f_{\mathrm{MB}}(E)
=
A\, e^{-E/k_BT},
$$
where the normalization constant satisfies  
$$
A^{-1}
=
\int_0^\infty e^{-E/k_BT}\, dE
=
k_BT.
$$
Thus the normalized form is  
$$
f_{\mathrm{MB}}(E)
=
\frac{1}{k_BT}\, e^{-E/k_BT}.
\qquad\blacksquare
$$
