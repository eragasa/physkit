# 1D Tight Binding: Splitting of an Atomic Level

Purpose:

Show how one isolated atomic energy level becomes several distinct energy levels when atomic orbitals are coupled.

Main model:

$$
H_{\mathrm{open}}(r)
=
E_0 I
-
t(r)
\sum_{n=0}^{N-2}
\left(
|n\rangle\langle n+1|
+
|n+1\rangle\langle n|
\right)
$$

with

$$
t(r)=t_0 e^{-\beta(r-r_0)}
$$

Boundary condition:

Open chain.

Physical object:

A finite chain or finite cluster of atoms.

Main computational task:

Construct $H_{\mathrm{open}}(r)$ for many values of $r$, diagonalize each matrix, and plot the eigenvalues $E_\alpha(r)$.

Main lesson:

Off-diagonal hopping lifts the degeneracy of the isolated atomic level.

The open-chain calculation shows the basic mechanism of band formation: coupling between atomic orbitals splits a degenerate atomic level.

The periodic-chain calculation adds the essential symmetry of a crystal. Once periodic boundary conditions are imposed, the eigenstates can be labeled by wave vector $k$, and the split levels become the tight-binding band

$$
E(k)=\epsilon_0-2t\cos(ka)
$$
