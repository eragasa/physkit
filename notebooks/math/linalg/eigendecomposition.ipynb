# Determining Eigenvalues and Eigenvectors

Let $M$ be a $2 \times 2$ matrix,
$$
M = \begin{bmatrix} 1 & 2 \\ 2 & 1\end{bmatrix}
$$
(a) Find the eigenvalues and eigenvectors of ${M}$
(b) Show for each eigenvalue and eigenpair
$$
  M|\lambda_i\rangle =\lambda_i|\lambda_i\rangle
$$

```python
import numpy as np        # loads the linear algebra algebra library
```

```python
# Define the matrix as an array of arrays
M = np.array([[1, 2], [2, 1]])

# Calculate eigenvalue and eigenvector pairs
eigenvalues, eigenvectors = np.linalg.eig(M)

# eigenvalues is just a regular array
print("eigenvalues:\t", eigenvalues)

# eigenvectors is an array of arrays where the the eigenvectors are columns
print("eigenvectors:\n", eigenvectors)

# select columns like this
print("eigenvector 1:\t", eigenvectors[:,0])
print("eigenvector 2:\t", eigenvectors[:,1])

# ---
# Part B
# To iterate correctly, we transpose eigenvectors so each 'row' is a vector
for i in range(len(eigenvalues)):
    eigval = eigenvalues[i]
    eigvec = eigenvectors[:, i]  # Get the i-th column

    M_times_eigvec = np.dot(M, eigvec)
    eigval_times_eigvec = eigval * eigvec
    is_equal = np.allclose(M_times_eigvec, eigval_times_eigvec)
    print(f"i: {i}")
    print(f"\teigval:         {eigval}")
    print(f"\teigvec:         {eigvec}")
    print(f"\tM*eigvec:       {M_times_eigvec}")
    print(f"\teigval*eigvec:  {eigval_times_eigvec}")
    print(f"\tis equal:       {is_equal}")
    print("-" * 30)
```
