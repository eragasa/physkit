```python
import numpy as np
import matplotlib.pyplot as plt

def square_wave_series(x, N):
    f = np.zeros_like(x)
    for n in range(1, 2*N, 2):  # odd harmonics only
        f += (4 / (np.pi * n)) * np.sin(n * x)
    return f

x = np.linspace(-np.pi, np.pi, 1000)
plt.plot(x, square_wave_series(x, 1), label='1 term')
plt.plot(x, square_wave_series(x, 5), label='5 terms')
plt.plot(x, square_wave_series(x, 20), label='20 terms')
plt.title('Fourier Series Approximation of Square Wave')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# define integration limits
L_low = np.min(x)
L_high = np.max(x)

a_0 = 0
a_n = np.array([1.])
b_n = np.array([1.])

print(np.shape(a_n))
print(np.shape(b_n))
assert np.shape(a_n) == np.shape(b_n)
def get_fourier_series(x: np.ndarray,
                       a_0: np.number, 
                       a_n: np.ndarray, 
                       b_n: np.ndarray,
                       L: np.number
                       ) -> np.ndarray:
    assert np.shape(a_n) == np.shape(b_n)

    sin_series = np.empty(shape = (sin_.shape[0],a_n.shape[0]))
    cos_series = np.empty(shape = (cos_.shape[0],a_n.shape[0]))
  
    sin_ = a_0/2 * np.ones(shape = x.shape)
    cos_ = a_0/2 * np.ones(shape = x.shape)

    sin_series[:,0] = sin_
    cos_series[:,0] = cos_

    for i in range(a_n.shape[0]):
        n = i+1
        sin_: np.ndarray = a_n * np.sin(n * np.pi * x / L)
        cos_: np.ndarray = b_n * np.cos(n * np.pi * x / L)
        sin_series[:,n] = sin_
        cos_series[:,n] = cos_

    return sin_series+cos_series, sin_series, cos_series



fig, ax = plt.subplots(3,1)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.sin(x)+np.cos(x))
```
