An Ising state is
$$
\mathbf{s} = (
  s_1,
  s_2,
  \ldots,
  s_N  
), \qquad
s_i \in \{-1,+1\}
$$

Then the state space is
$$\mathcal{X} = \{-1,+1\}^N$$



```python
from dataclasses import dataclass
import numpy as np

from numpy.typing import NDArray

@dataclass(frozen=True)
class IsingState:
    spins: NDArray[np.int_]

@dataclass(frozen=True)
class IsingStateSpace:
    n_sites: int

    def contains(self, state: IsingState) -> bool:
        if state.spins.shape != (self.n_sites,):
            return False
        
        valid_spin_values = (state.spins == -1 | (state.spins == 1))
        return bool(np.all(valid_spin_values))
```

```python
state_space = IsingStateSpace(n_sites=4)
state = IsingState(
    spins=np.array([1, -1, 1, 1])
)

assert isinstance(state_space.contains(state), bool)
```
