## Lattice Vectors

$$\begin{gather}
\mathbf{a}_1 = (a_{11}, a_{12}, a_{13}) \\
\mathbf{a}_2 = (a_{21}, a_{22}, a_{23}) \\
\mathbf{a}_3 = (a_{31}, a_{32}, a_{33})
\end{gather}$$

```python
import numpy as np

a11, a12, a13 = (12,12,13)
a21, a22, a23 = (21,22,23)
a31, a32, a33 = (31,32,33)

a1 = np.array([a11, a12, a13], dtype=float)
a2 = np.array([a21, a22, a23], dtype=float)
a3 = np.array([a31, a32, a33], dtype=float)

print(f"a1:{a1}")
print(f"a2:{a2}")
print(f"a3:{a3}")

A = np.vstack([a1, a2, a3]).T
print(f"A:{A}")
```
