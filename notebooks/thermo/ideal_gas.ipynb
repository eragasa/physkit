```python
import numpy as np
import matplotlib.pyplot as plt
# PV = nRT

class IdealGas():
    @staticmethod
    def calc_pressure(V, T, n=1., R=1.):
        P = n*R*T/V
        return P
    
fig, ax = plt.subplots(nrows=1, ncols=1)

V = np.linspace(1e-5,2,100)
T_arr = [.01, .02, .05, .1, .2, .5]

for T in T_arr:
  P = IdealGas.calc_pressure(V, T)
  ax.plot(
     V,P,label=f"T={T:.2f}"
  )
ax.legend()
ax.set_xlim(0,2)
ax.set_ylim(0,2)
ax.set_xlabel('P')
ax.set_ylabel('V')
plt.show()
```
