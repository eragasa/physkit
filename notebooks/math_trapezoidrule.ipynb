# Visual Example: Trapezoidal Rule

We approximate the integral
$$
I = \int_0^1 e^{-x^2}\,dx
$$
using the trapezoidal rule with $N=6$ subintervals.  
The plot shows the function $f(x)=e^{-x^2}$ and the trapezoids used in the approximation.

```python

````python

import numpy as np
import matplotlib.pyplot as plt

# function
def func_exp(x, alpha=2):
    return np.exp(-x**2)

f = func_exp

# interval and resolution
x_min, x_max = 0,1
N_x = 30

# integrate trapezoid rule
x = np.linspace(x_min, x_max, N_x+1)
y = f(x)
h = (x_max - x_min) / N_x

# trapezoidal approximation
T = h * 0.5*y[0] 
T += np.sum(y[1:-1])
T += 0.5*y[-1]

N_x_highres = 400
# high-resolution curve for plotting
x_min_highres = x_min - 0.1
x_max_highres = x_max + 0.1
x_highres = np.linspace(x_min_highres, x_max_highres, N_x_highres)
y_highres = f(x_highres)

plt.figure(figsize=(4,4))

# plot the smooth function
plt.plot(x_highres, y_highres, label="f(x) = exp(-x^2)", linewidth=2)

# plot trapezoids
for i in range(N_x):
    xs = [x[i], x[i], x[i+1], x[i+1]]
    ys = [0, y[i], y[i+1], 0]
    plt.fill(xs, ys, alpha=0.3, edgecolor='k', color='blue')

# grid and labels
plt.scatter(x, y, color='k')
plt.vlines(x, 0, y, colors='gray', linestyle='--', linewidth=0.6)
plt.title(f"Trapezoidal Rule Approximation (N={N_x})\nT â‰ˆ {T:.6f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.tight_layout()
plt.show()
````
