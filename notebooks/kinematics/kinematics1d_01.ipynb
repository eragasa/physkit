The study of kinematics is the study of 
$$
  x(t), v(t), a(t), t
$$
where
- $x(t)$: is the position of the particle at time $t$
- $v(t)$: is the velocity of the particle at time $t$
- $a(t)$: is the acceleration of the particle at time $t$
- $t$: is the time coordinate
But since 
$$
  v(t)
  =\frac{\mathrm{d}x}{\mathrm{d}t}
  \approx \frac{\Delta x}{\Delta t}
  =\frac{x_f-x_i}{t_f-x_i}
$$
All our motion happens on $x\in(-\infty,\infty)$ with $x \in \mathbb{R}$

Computers typically do computations through discretization, that is we are limited to a finite domain $x\in(-x_\text{min},x_\text{max})$ with an arbitrary number of samples sampled at identical intervals.

Let's say we want to model
$$
  x=x_0+v\Delta t
$$
then $t$ is a coordinate system but we don't know $t_f$ but know $t_0$ and $x_0$



````python
import numpy as np
x_0 = 0
t_0 = 0
t_f = 5.
v = 1.
N = 100

t_arr = np.linspace(t_0, t_f, 100)
dt = t_arr[1]-t_arr[0]
print(f"dt={dt:4f}")
v_arr = v * np.ones(shape=t_arr.shape)
print(f"v_arr=",v_arr)

x_arr = np.zeros(shape=t_arr.shape)
for i in range(1,t_arr.shape[0]):
  dx = v_arr[i-1]*dt
  x_arr[i] = x_arr[i-1] + dx
  print(i, t_arr[i-1], dx, x_arr[i])
print(f"x_arr=",x_arr)


````

````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,1)
fig.set_size_inches(3,3)

ax[0].plot(t_arr, v_arr)
ax[1].plot(t_arr, x_arr)
fig.tight_layout()
````
