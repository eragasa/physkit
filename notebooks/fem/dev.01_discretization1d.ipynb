Let us discretize $x \in [0,L]$ over $N_x$ points

Then $x_\text{min}=0$ and $x_\text{max=L}$


```python
L = 1.
N_x = 10

x_min = 0.
x_max = x_min + L

# example on printing out the variables
print(f"L={L}")
print(f"N_x={N_x}")
print(f"x_min={x_min}")
print(f"x_max={x_max}")
```

```python
import numpy as np
x_arr = np.linspace(
    start = x_min,
    stop = x_max,
    num = N_x
)

# commenting the keywords is optional, but when I finish writing code it is
# preferrable to to put in the keywords
x_arr_2 = np.linspace(x_min, x_max, N_x)

# are they the same?
print("x_arr == x_arr_2? {}".format(np.allclose(x_arr, x_arr_2)))

# which is the same as
all_equal = True
for x1, x2 in zip(x_arr, x_arr_2):
  if x1 != x2:
    all_equal = False
print("x_arr == x_arr_2? {}".format(all_equal))


print("x_arr\n",x_arr)
```

```python
print("x_arr.shape:{}".format(x_arr.shape))
print()
for i in range(x_arr.shape[0]):
  print("{} {:.4f}".format(i,x_arr[i]))

# Alternatively
print()
for i in range(x_arr.shape[0]):
  print(f"{i} {x_arr[i]:.4f}")
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.scatter(
  x=x_arr,
  y=np.zeros(shape=x_arr.shape)
)
```
