## Antoine Equation

$$
  \log_{10}(P)=A-(B/(T+C))
$$

```python
import numpy as np

class AntoineEquation():
  def __init__(self, A: float, B: float, C: float, T_low: float, T_high: float):
    self.T_low: float = T_low
    self.T_high: float = T_high
    self.A: float = A
    self.B: float = B
    self.C: float = C

  @staticmethod
  def calculate_pressure_bars(T, A, B, C):
    logP = A - (B/T+C)
    P_bar = 10**logP
    return P_bar
  
  def get_pressure_bars(self, T):
    P_bar = AntoineEquation.calculate_pressure_bars(
      T = T,
      A = self.A,
      B = self.B,
      C = self. C
    )
    return P_bar
```

```python
from dataclasses import dataclass

@dataclass
class JanafNistAntoineParameters:
  T_low: float
  T_high: float
  A: float
  B: float
  C: float


class JanafNistInformation():
  antoine_parameters: JanafNistAntoineParameters
AlJanafNistInformation = JanafNistInformation
antoine_parameters = {
  "T_low": 1557.,
  "T_high": 2329.,
  "A": 5.73623,
  "B": 13204.109,
  "C": -24.306	
}

# https://webbook.nist.gov/cgi/cbook.cgi?ID=C7429905&Mask=4&Plot=on&Type=ANTOINE&utm_source=chatgpt.com



Al_vapor_curve = AntoineEquation(**Al_antoine_parameters)

T_arr_kelvin = np.linspace(Al_vapor_curve.T_low, Al_vapor_curve.T_high, 1000)
P_arr_bars = Al_vapor_curve.get_pressure_bars(T_arr_kelvin)

import matplotlib.pyplot as plt
plt.plot(T_arr_kelvin,P_arr_bars)
```
