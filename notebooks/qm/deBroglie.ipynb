#### Relativistic factor
$$
\gamma = \frac{1}{\sqrt{1-v^2/c^2}}
$$

#### Length Contraction
$$
L = \gamma L_0
$$

#### Time Dialation
$$
t = \gamma t_0
$$

#### Relativistic momentum
$$
\mathbf{p}=\gamma m \mathbf{v}
$$

#### Relativistic 2nd Law
$$\begin{align}
\mathbf{F} 
&= \frac{\mathrm{d}\mathbf{p}}{\mathrm{d}t} \\
&= \frac{\mathrm{d}}{\mathrm{d}t}(\gamma m \mathbf{v})
\end{align}$$

```python
from physkit.constants import ConstantsSI
import numpy as np

class SpecialRelativity:
    @staticmethod
    def calculate_gamma(v: float, c: float = ConstantsSI.c):
        gamma: float = 1 / np.sqrt(1-(v**2/c**2))
        return gamma

class LengthContraction:
    @staticmethod
    def calculate_L(L0: float, v: float, c: float = ConstantsSI.c):
        gamma = SpecialRelativity.calculate_gamma(v=v, c=c)
        L = L0 * gamma
        return L
        
class TimeDilation:
    @staticmethod
    def calculate_t(t0: float, v: float, c: float = ConstantsSI.c):
        """
        Arguments:
        t0: float
          time interval on clock at rest relative to an observer, proper time
        v: float
          speed of relative motion, same unit system as `c`, default m/s
        c: float
          speed of light, default SI units

        Returns:
        (float) time interval on clock in motion relative to an observer
        """
        gamma = SpecialRelativity.calculate_gamma(v=v, c=c)
        t = gamma * t0
        return t

class RelativisiticMomentum:
    @staticmethod
    def calculate_p(m: float, v: np.ndarray, c=ConstantsSI.c):
        """
        m (float)
          the proper mass (rest mass) of an object
        v (float)
          
        v
        """
        gamma = SpecialRelativity.calculate_gamma(
            v=np.sqrt(np.dot(v,v)),
            c=c
        )
        p = gamma * m * v
        return v
```

#### Adapted from Besier, 6th Ed, Example 1.1
A spacecraft is moving relative to the earth.  An observer on the earth finds that between 1 P.M. and 2 P.M. according to her clock, $3601\text{ s}$ elapse on the spacecraft's clock.  What is the spacecraft's speed relative to the earth

The proper time interval on earth is the difference on hour


```python
from physkit.units import Time

t0 = Time.convert(value=1, 
                  units_from=Time.Units.hr, 
                  units_to=Time.Units.s)
print(f"t0={t0} seconds")
t = 3601

from scipy.optimize import brentq
def f(v, t0, t):
    return t-TimeDilation.calculate_t(t0=t0, v=v, c=ConstantsSI.c)

root, result = brentq(
    f,
    a=0,
    b=ConstantsSI.c - 1,
    args=(t0, t),
    full_output=True
)

# ------------------------------------------------------------
# Print solver diagnostics
# ------------------------------------------------------------
print(f"root            = {root:.4e} s")
print(f"converged       = {result.converged}")
print(f"iterations      = {result.iterations}")
print(f"function calls  = {result.function_calls}")
print(f"f({root:.4e})   = {f(root, t0, t)}")
```

# DeBroglie Waves


```python
from physkit.constants import ConstantsSI
import numpy as np

    

class DeBroglieWave:
    @staticmethod
    def calculate_wavelength(
        m: float, 
        v: float, 
        gamma: float=1., 
        h: float=ConstantsSI.h
    ) -> float:
        """
        m (float):
            mass in units of `h`, by default kg
        v (float):
            velocity in units of `h`, by default `m/s`
        gamma (float):
            relativistic factor, in the units of `h`, but is n.d.
        h (float):
            planck's constant, by default using SI units
        """
        lambda_: float = h/(gamma*m*v)
        return lambda_
    @staticmethod
    def calculate_wave_velocity(nu_: float, lambda_: float):
        """
        lambda_ (float):
            de Broglie wavelength
        """

        nu_p = nu_ * lambda_
        return nu_p
```

## Problem.   Beiser, 6th Ed, Example 6.1

Find the de Broglie wavelength of 
(a) a 46 $\text{g}$ golf ball with a velocity of $30\text{ m/s}$ 
(b) an electron with a velocity of $10^7\text{ m/s}$

```python
from physkit.units import Mass
## part a
m = Mass.convert(from_=[46, Mass.Units.g], to=Mass.Units.kg)
v = 30
gamma = SpecialRelativity.calculate_gamma(v=v,c=ConstantsSI.c)
print(f"m={m} kg")
print(f"v={v} m/s")
print(f"gamma={gamma} [n.p.]")
debroglie_lambda = DeBroglieWave.calculate_wavelength(
    m=m,
    v=v,
    gamma=gamma,
    h=ConstantsSI.h
)
print(f"lambda={debroglie_lambda}")

```

```python
## part b
m = ConstantsSI.me0
v = 1.e7
gamma = SpecialRelativity.calculate_gamma(v=v,c=ConstantsSI.c)
print(f"m={m} kg")
print(f"v={v} m/s")
print(f"gamma={gamma} [n.p.]")
print(f"h={ConstantsSI.h} J s")

debroglie_lambda_nonrelatistic = DeBroglieWave.calculate_wavelength(
    m=m,
    v=v,
    gamma=1.,
    h=ConstantsSI.h
)
debroglie_lambda_relatistic = DeBroglieWave.calculate_wavelength(
    m=m,
    v=v,
    gamma=gamma,
    h=ConstantsSI.h
)
print(f"lambda={debroglie_lambda_nonrelatistic:.4e}")
print(f"lambda={debroglie_lambda_relatistic:.4e}")
```
