```python
import numpy as np

def get_vonmises_stress_3D(sigma_tensor):
    """
    Compute the 3D von Mises equivalent stress from a stress tensor.

    Assumes a 3D Cauchy stress tensor of the form:

        [[sigma_xx, tau_xy, tau_xz],
         [tau_yx, sigma_yy, tau_yz],
         [tau_zx, tau_zy, sigma_zz]]

    For a symmetric stress tensor:

        tau_xy = tau_yx
        tau_xz = tau_zx
        tau_yz = tau_zy

    The 3D von Mises stress is:

        sigma_vm = sqrt(
            0.5 * (
                (sigma_xx - sigma_yy)^2
              + (sigma_yy - sigma_zz)^2
              + (sigma_zz - sigma_xx)^2
            )
            + 3 * (tau_xy^2 + tau_yz^2 + tau_xz^2)
        )
    """

    sigma_tensor = np.asarray(sigma_tensor, dtype=float)

    if sigma_tensor.shape != (3, 3):
        raise ValueError("sigma_tensor must be a 3x3 array.")

    sigma_xx = sigma_tensor[0, 0]
    sigma_yy = sigma_tensor[1, 1]
    sigma_zz = sigma_tensor[2, 2]

    # Average off-diagonal terms in case of small numerical asymmetry
    tau_xy = 0.5 * (sigma_tensor[0, 1] + sigma_tensor[1, 0])
    tau_xz = 0.5 * (sigma_tensor[0, 2] + sigma_tensor[2, 0])
    tau_yz = 0.5 * (sigma_tensor[1, 2] + sigma_tensor[2, 1])

    sigma_vm = np.sqrt(
        0.5 * (
            (sigma_xx - sigma_yy)**2
            + (sigma_yy - sigma_zz)**2
            + (sigma_zz - sigma_xx)**2
        )
        + 3.0 * (
            tau_xy**2
            + tau_yz**2
            + tau_xz**2
        )
    )

    return sigma_vm
```
