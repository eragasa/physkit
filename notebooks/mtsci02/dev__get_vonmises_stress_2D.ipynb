### Function: `get_vonmises_stress_2D`

Computes the von Mises equivalent stress for a two-dimensional stress tensor.

This function assumes a plane-stress tensor of the form

$$
\boldsymbol{\sigma}
=
\begin{bmatrix}
\sigma_{xx} & \tau_{xy} \\
\tau_{yx} & \sigma_{yy}
\end{bmatrix}.
$$

For a physically symmetric Cauchy stress tensor,

$$
\tau_{xy} = \tau_{yx}.
$$

The function averages the two off-diagonal terms,

$$
\tau_{xy}
=
\frac{1}{2}
\left(
\sigma_{xy} + \sigma_{yx}
\right),
$$

to reduce the effect of small numerical asymmetries.

The von Mises equivalent stress in 2D plane stress is

$$
\sigma_\mathrm{vM}
=
\sqrt{
\sigma_{xx}^2
-
\sigma_{xx}\sigma_{yy}
+
\sigma_{yy}^2
+
3\tau_{xy}^2
}.
$$

#### Parameters

| Parameter | Type | Description |
|---|---:|---|
| `sigma_tensor` | array-like, shape `(2, 2)` | Two-dimensional stress tensor. Entries should be in consistent stress units, such as Pa, MPa, or psi. |

#### Returns

| Return value | Type | Description |
|---|---:|---|
| `sigma_vm` | `float` | Von Mises equivalent stress in the same units as the input stress tensor. |

#### Raises

| Error | Cause |
|---|---|
| `ValueError` | Raised if `sigma_tensor` is not a `2 x 2` array. |

#### Example

```python
import numpy as np

sigma = np.array([
    [100.0, 25.0],
    [25.0,  40.0]
])

sigma_vm = get_vonmises_stress_2D(sigma)

print(sigma_vm)

````python
import numpy as np

def get_vonmises_stress_2D(sigma_tensor):
    """
    Compute the 2D von Mises stress from a stress tensor.

    Assumes a 2D stress tensor of the form:

        [[sigma_xx, tau_xy],
         [tau_yx,   sigma_yy]]

    For plane stress:

        sigma_vm = sqrt(sigma_xx^2 - sigma_xx*sigma_yy + sigma_yy^2 + 3*tau_xy^2)
    """

    sigma_tensor = np.asarray(sigma_tensor, dtype=float)

    if sigma_tensor.shape != (2, 2):
        raise ValueError("sigma_tensor must be a 2x2 array.")

    sigma_xx = sigma_tensor[0, 0]
    sigma_yy = sigma_tensor[1, 1]

    # Average the off-diagonal terms in case of small numerical asymmetry
    tau_xy = 0.5 * (sigma_tensor[0, 1] + sigma_tensor[1, 0])

    sigma_vm = np.sqrt(
        sigma_xx**2
        - sigma_xx * sigma_yy
        + sigma_yy**2
        + 3.0 * tau_xy**2
    )

    return sigma_vm
````
