$F_x = - kx$

```python
# plotting the spring constant

import numpy as np
x_min = -2.
x_max = 2.
N_x = 100

k = 1.   # spring constant, N/m
k_units = "N/m"
x_arr = np.linspace(x_min, x_max, N_x)

def hooks_law(k: float, x):
    F_spring = -k * x 
    return F_spring

Fx_arr = hooks_law(k, x=x_arr)

import matplotlib.pyplot as plt

# create figure and axes
fig, ax = plt.subplots(1,1)
fig.set_size_inches(4,4)

# draw horizontal line at x=0
ax.axhline(0, color="grey", linewidth=1)
# draw vertical line at y=0
ax.axvline(0, color="grey", linewidth=1)


ax.plot(x_arr,Fx_arr)

# add titles
title = f"Hook's Law $(k={k:.2f}\\,\\text{{{k_units}}})$"
x_label = "position [m]"
y_label = "$F_s$ - spring force [N/m]"
ax.set_title(title)
ax.set_xlabel(x_label)
ax.set_ylabel(y_label)

ax.grid(visible=False)

```
