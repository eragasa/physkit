```python
from abc import ABC, abstractmethod
from dataclasses import dataclass

import numpy as np
import scipy.sparse


ArrayFloat64 = np.typing.NDArray[np.float64]


class QuantumPotential1D(ABC):
    """
    Base class for a local scalar quantum potential represented
    on a specific 1D position grid.
    """

    x: ArrayFloat64

    @abstractmethod
    def values(self) -> ArrayFloat64:
        raise NotImplementedError

    def matrix(self) -> scipy.sparse.csr_matrix:
        return scipy.sparse.diags(self.values(), offsets=0, format="csr")
```

```python
@dataclass(frozen=True)
class ParticleInABoxPotential1D(QuantumPotential1D):
    x: ArrayFloat64
    length: float = len()
    origin: float = x[0]

    def values(self) -> ArrayFloat64:
        return np.zeros_like(self.x, dtype=float)
```
